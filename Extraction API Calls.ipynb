```python
# combining all of the above into functions

api_key = 'AIzaSyCbVEbw6azQEqhoOe3CzTBqjdv4OJv8M8Q'

def get_duration(dur):
    if 'H' in dur:
        hours = dur[dur.index('T') + 1 : dur.index('H')]
        minutes = dur[dur.index('H') + 1 :dur.index('M')]
        seconds = dur[dur.index('M') + 1 : dur.index('S')]
        minutes = minutes + "." + seconds
        minutes = float(minutes) + float(hours)*60
        total_duration = int(minutes)

    else:
        minutes = dur[dur.index('T') + 1 :dur.index('M')]
        seconds = dur[dur.index('M') + 1 : dur.index('S')]

        minutes = minutes + '.' + seconds

        total_duration = float(minutes)
        total_duration = int(total_duration)


    return total_duration
    

def get_video_stats(video_id):
    url_vs = "https://www.googleapis.com/youtube/v3/videos?id=" + video_id + "&part=snippet,statistics,contentDetails,player&key=" + api_key
    response_vs = requests.get(url_vs).json()
    vid_desc = response_vs['items'][0]['snippet']['description']
#     player = response_vs['items'][0]['player']['embedHtml'].split(' ')[3].split('"')[1].replace('//', '')
    player =  response_vs['items'][0]['player']
    duration = response_vs['items'][0]['contentDetails']['duration']
#     duration = get_duration(dur)
    
    try: # encountered a key erroe, hence the try-catch
        views = int(response_vs['items'][0]['statistics']['viewCount'])
        likes = int(response_vs['items'][0]['statistics']['likeCount'])
        dislikes = int(response_vs['items'][0]['statistics']['dislikeCount'])
        if views > 0:
            like_perc = likes / views 
        else:
            like_perc = 0  # got a ZeroDivisionError 
    #     fav = response_vs['items'][0]['statistics']['favouriteCount']
        comments = int(response_vs['items'][0]['statistics']['commentCount'])
        
        if 'tags' in response_vs['items'][0]['snippet'].keys(): # video description may not contain tags always
            tags = response_vs['items'][0]['snippet']['tags']
        
        else:
            tags = []
        
    except KeyError:
        print("Key not found!")
        views, likes, dislikes, like_perc, comments, tags, vid_desc = np.NaN, np.NaN, np.NaN, np.NaN, np.NaN, np.NaN, np.NaN 
    
    return views, likes, dislikes, like_perc, comments, tags, vid_desc, duration, player 

def get_subs_count(channel_id):
    url_subs = 'https://www.googleapis.com/youtube/v3/channels?key=' + api_key + '&part=statistics&id=' + channel_id
    response_subs = requests.get(url_subs).json()
    subs = response_subs['items'][0]['statistics']['subscriberCount']
    
    return subs 

def get_playlist(channel_id, playlist_id):
    url_pl = "https://www.googleapis.com/youtube/v3/playlists?key=" + api_key + "&channelId=" + channel_id + "&id=" + playlist_id + "&part=contentDetails,snippet,player"


def get_videos(channel_id, df):
    pageToken = ""
    while 1:
        url = "https://www.googleapis.com/youtube/v3/search?key=" + api_key + "&channelId=" + channel_id + "&part=snippet,id&order=date&maxResults=10000&"+pageToken
        response = requests.get(url).json()
        time.sleep(1) # giving it a second before starting the for loop
        
        try:
            for video in response['items']:
                if video['id']['kind'] == "youtube#video": # making sure its a video
                    channel_name = video['snippet']['channelTitle']
                    vid_id = video['id']['videoId']
                    vid_title = video['snippet']['title']
                    pub_date = video['snippet']['publishTime']
                    subs = get_subs_count(channel_id)
                    views, likes, dislikes, like_perc, comments, tags, desc, duration, player = get_video_stats(vid_id)
                    
                     # creating dataframe
                    df = df.append({'channel': channel_name, 'id':vid_id,'title':vid_title, "upload_date": pub_date, 'subscriber_count': subs, "view_count":views, "like_count":likes,"dislike_count":dislikes,
                                "like_perc": like_perc, "comment_count":comments, 'tags': tags, 'description': desc, 'duration': duration, 'player': player}, ignore_index=True, )
                    print(len(df))

                    
        except KeyError:
            continue

               
        try:
            if response['nextPageToken'] != None: # if none, means we have reached the last page
                pageToken = 'pageToken=' + response['nextPageToken']
    #             get_videos(channel_id, df)
        except:
            break
    
    return df
```
